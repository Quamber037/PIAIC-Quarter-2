```python
def resize(files, Dir):  
    counter = 0
    pic = 1
   
    for i in files: 
        list = os.listdir(".\photos")
        number_files = len(list)
        arr = np.zeros([number_files,200,200,4])
        print("PIC NO:",pic)
        img = PIL.Image.open(Dir+"//"+i)
        n_img = img.resize((200,200))
        arr[counter] = n_img
        print("Image Resized:")
        print("Shape :",arr.shape)
        print("Size :",arr.size)
        print("\n")
        pic+=1
        counter+=1
        
def Apply():  #Searching for files   
    resize(files, Dir)
```

```python
if __name__== "__main__":
    Apply()
```
